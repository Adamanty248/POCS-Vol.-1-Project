```python
def sentiment_analyzer(words):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer 
    import numpy as np
    import re    #used for splitting text - re.split()
    import matplotlib.pyplot as plt
    Title = 'On the Brink of Separation'

    outfile = open('Text_Files\On the Brink of Separation.txt','w')
    outfile.write(words)
    outfile.close()

    words = re.sub('”','',words)
    words = re.sub('“','',words)
    words = re.sub('\?','\.',words)

    sentiment = []
    list_of_sentences = re.split(str('\.'),words)
    for sentence in list_of_sentences:
        sid = SentimentIntensityAnalyzer()
        #print(sentence)
        ss = sid.polarity_scores(sentence)
        #for k in sorted(ss):
            #print('{0}: {1}, '.format(k, ss[k]), end='')
        #print()
        sentiment.append(ss['compound'])
        #print(' ')


    #print(sentiment)
    sentiment_of_sentences =np.zeros(len(sentiment))
    sentences = np.zeros(len(sentiment))
    init = 0
    for j,i in enumerate(sentiment):
        init += i
        sentiment_of_sentences[j] = init
        sentences[j] = j+1
    sentences = np.divide(sentences, len(sentences))
    plt.style.use('ggplot')
    fig,ax = plt.subplots(figsize=(5,5))
    ax.set(title = 'On the Brink of Separation',ylabel ='Sentiment',xlabel = 'Proportion of Sentences')
    ax.plot(sentences, sentiment_of_sentences)
    plt.savefig('Sentiment_Plots/{}' .format(Title))

    #plt.show()
sentiment_analyzer(words)
```
